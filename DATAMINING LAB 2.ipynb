```python
def equifreq(d,m):
    
    d.sort()
    a = len(d)
    n = int(a/m)
    bins = []
    
    for i in range(m):
        bin = []
        
        if i == m - 1:
            end_index = a
        else:
            end_index = (i+1)*n
        
        for j in range(i*n, end_index):
            bin.append(d[j])
        bins.append(bin)
    
    
    print("Original Bins:")
    for b in bins:
        print(b)
    
    
    print("\nSmoothed by Mean:")
    for b in bins:
        mean_val = sum(b) / len(b)
        print([mean_val] * len(b))
    
    
    print("\nSmoothed by Bin Boundaries:")
    smoothed_bins = []
    for b in bins:
        # Define min and max inside the loop where 'b' is available
        min_val, max_val = min(b), max(b)
        smoothed = []
        for val in b:
            if abs(val - min_val) <= abs(val - max_val):
                smoothed.append(min_val)
            else:
                smoothed.append(max_val)
        smoothed_bins.append(smoothed)
    
    for b in smoothed_bins:
        print(b)
       

data = [11, 6, 7, 8, 24, 55, 66, 77, 89]
m = 3
equifreq(data, m)
```
